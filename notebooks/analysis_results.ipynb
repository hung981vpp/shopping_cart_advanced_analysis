```python
import pandas as pd
import numpy as np

print("="*70)
print("üìä PH√ÇN T√çCH K·∫æT QU·∫¢ LAB 2 - MIN_SUPPORT=0.02")
print("="*70)

# 1. Load c·∫£ 2 b·ªô rules (th√™m ../ v√¨ ƒëang ·ªü trong notebooks/)
rules_apriori = pd.read_csv('../data/processed/rules_apriori_filtered.csv')
rules_fpgrowth = pd.read_csv('../data/processed/rules_fpgrowth_filtered.csv')

# 2. So s√°nh s·ªë l∆∞·ª£ng
print("\nüî¢ SO S√ÅNH S·ªê L∆Ø·ª¢NG:")
print(f"‚îú‚îÄ Apriori:    {len(rules_apriori):,} lu·∫≠t")
print(f"‚îú‚îÄ FP-Growth:  {len(rules_fpgrowth):,} lu·∫≠t")
print(f"‚îî‚îÄ Ch√™nh l·ªách: {abs(len(rules_apriori) - len(rules_fpgrowth))} lu·∫≠t")

if len(rules_apriori) == len(rules_fpgrowth):
    print("   ‚úÖ GI·ªêNG NHAU - ƒê√∫ng l√Ω thuy·∫øt!")
else:
    print("   ‚ö†Ô∏è KH√ÅC NHAU - C·∫ßn ki·ªÉm tra l·∫°i!")

# 3. Th·ªëng k√™ Apriori
print("\n" + "="*70)
print("üìà TH·ªêNG K√ä APRIORI")
print("="*70)
print("\nSupport:")
print(rules_apriori['support'].describe().to_string())
print("\nConfidence:")
print(rules_apriori['confidence'].describe().to_string())
print("\nLift:")
print(rules_apriori['lift'].describe().to_string())

# 4. Top 5 rules theo Lift
print("\n" + "="*70)
print("üèÜ TOP 5 RULES THEO LIFT (APRIORI)")
print("="*70)

# Ki·ªÉm tra xem c√≥ c·ªôt 'antecedents_str' kh√¥ng
if 'antecedents_str' in rules_apriori.columns:
    cols = ['antecedents_str', 'consequents_str', 'support', 'confidence', 'lift']
else:
    cols = ['antecedents', 'consequents', 'support', 'confidence', 'lift']

top5_apriori = rules_apriori.nlargest(5, 'lift')[cols]
for idx, row in top5_apriori.iterrows():
    print(f"\n{idx+1}. {row[cols[0]]} ‚Üí {row[cols[1]]}")
    print(f"   Support: {row['support']:.4f} | Confidence: {row['confidence']:.4f} | Lift: {row['lift']:.2f}")

# 5. Ph√¢n b·ªë Lift
print("\n" + "="*70)
print("üìä PH√ÇN B·ªê LIFT")
print("="*70)
lift_bins = pd.cut(rules_apriori['lift'], bins=[0, 1, 2, 3, 5, 10, np.inf])
print(rules_apriori.groupby(lift_bins).size().to_string())

# 6. Basket info
basket = pd.read_parquet('../data/processed/basket_bool.parquet')
print("\n" + "="*70)
print("üõí TH√îNG TIN BASKET")
print("="*70)
print(f"S·ªë invoices:  {len(basket):,}")
print(f"S·ªë s·∫£n ph·∫©m:  {len(basket.columns):,}")
print(f"Density:      {(basket.sum().sum() / (len(basket) * len(basket.columns))):.2%}")

# 7. Top 10 s·∫£n ph·∫©m ph·ªï bi·∫øn nh·∫•t
print("\n" + "="*70)
print("üî• TOP 10 S·∫¢N PH·∫®M PH·ªî BI·∫æN NH·∫§T")
print("="*70)
top_products = basket.sum().nlargest(10)
for rank, (product, count) in enumerate(top_products.items(), 1):
    pct = count / len(basket) * 100
    print(f"{rank:2d}. {product[:50]:50s} | {count:5d} ({pct:5.2f}%)")

print("\n" + "="*70)
print("‚úÖ PH√ÇN T√çCH HO√ÄN T·∫§T!")
print("="*70)
```

```python
import pandas as pd

# Load 2 b·ªô rules
rules_ap = pd.read_csv('../data/processed/rules_apriori_filtered.csv')
rules_fp = pd.read_csv('../data/processed/rules_fpgrowth_filtered.csv')

print("="*70)
print("üîç SO S√ÅNH CHI TI·∫æT")
print("="*70)

# 1. Th·ªëng k√™ c∆° b·∫£n
print("\nüìä APRIORI (175 rules):")
print(rules_ap[['support', 'confidence', 'lift']].describe())

print("\nüìä FP-GROWTH (1,794 rules):")
print(rules_fp[['support', 'confidence', 'lift']].describe())

# 2. Ph√¢n b·ªë Confidence
print("\n" + "="*70)
print("üìà PH√ÇN B·ªê CONFIDENCE")
print("="*70)
print(f"Apriori min confidence:   {rules_ap['confidence'].min():.4f}")
print(f"Apriori median confidence: {rules_ap['confidence'].median():.4f}")
print(f"\nFP-Growth min confidence:   {rules_fp['confidence'].min():.4f}")
print(f"FP-Growth median confidence: {rules_fp['confidence'].median():.4f}")

# 3. Ph√¢n b·ªë Lift
print("\n" + "="*70)
print("üìà PH√ÇN B·ªê LIFT")
print("="*70)
print(f"Apriori min lift:   {rules_ap['lift'].min():.4f}")
print(f"Apriori median lift: {rules_ap['lift'].median():.4f}")
print(f"\nFP-Growth min lift:   {rules_fp['lift'].min():.4f}")
print(f"FP-Growth median lift: {rules_fp['lift'].median():.4f}")

# 4. Test: N·∫øu l·ªçc FP-Growth gi·ªëng Apriori th√¨ c√≤n bao nhi√™u?
print("\n" + "="*70)
print("üß™ TH·ª∞C NGHI·ªÜM: L·ªåC FP-GROWTH GI·ªêNG APRIORI")
print("="*70)

min_conf_ap = rules_ap['confidence'].min()
min_lift_ap = rules_ap['lift'].min()

fp_filtered = rules_fp[
    (rules_fp['confidence'] >= min_conf_ap) &
    (rules_fp['lift'] >= min_lift_ap)
]

print(f"FP-Growth g·ªëc:                {len(rules_fp):,} rules")
print(f"Apriori:                       {len(rules_ap):,} rules")
print(f"FP-Growth sau khi l·ªçc gi·ªëng:  {len(fp_filtered):,} rules")
print(f"\n‚û°Ô∏è Ch√™nh l·ªách c√≤n l·∫°i:        {abs(len(fp_filtered) - len(rules_ap)):,} rules")

if abs(len(fp_filtered) - len(rules_ap)) < 10:
    print("\n‚úÖ SAU KHI L·ªåC GI·ªêNG NHAU ‚Üí V·∫•n ƒë·ªÅ l√† filter kh√°c nhau!")
else:
    print("\n‚ö†Ô∏è V·∫™N KH√ÅC NHAU ‚Üí C√≥ v·∫•n ƒë·ªÅ kh√°c (c√≥ th·ªÉ antecedent length)")

# 5. Ki·ªÉm tra antecedent length
print("\n" + "="*70)
print("üìè PH√ÇN B·ªê ƒê·ªò D√ÄI ANTECEDENTS")
print("="*70)

# C·∫ßn parse antecedents n·∫øu l√† string
if isinstance(rules_ap.iloc[0]['antecedents'], str):
    rules_ap['ante_len'] = rules_ap['antecedents'].str.count(',') + 1
    rules_fp['ante_len'] = rules_fp['antecedents'].str.count(',') + 1
else:
    rules_ap['ante_len'] = rules_ap['antecedents'].apply(len)
    rules_fp['ante_len'] = rules_fp['antecedents'].apply(len)

print("Apriori antecedent length:")
print(rules_ap['ante_len'].value_counts().sort_index())

print("\nFP-Growth antecedent length:")
print(rules_fp['ante_len'].value_counts().sort_index())

print("\n" + "="*70)
```

```python
import pandas as pd

print("="*70)
print("‚úÖ KI·ªÇM TRA K·∫æT QU·∫¢ SAU KHI S·ª¨A")
print("="*70)

# Load 2 b·ªô rules
rules_ap = pd.read_csv('../data/processed/rules_apriori_filtered.csv')
rules_fp = pd.read_csv('../data/processed/rules_fpgrowth_filtered.csv')

# 1. So s√°nh s·ªë l∆∞·ª£ng
print("\nüî¢ SO S√ÅNH S·ªê L∆Ø·ª¢NG:")
print(f"‚îú‚îÄ Apriori:    {len(rules_ap):,} rules")
print(f"‚îú‚îÄ FP-Growth:  {len(rules_fp):,} rules")
print(f"‚îî‚îÄ Ch√™nh l·ªách: {abs(len(rules_ap) - len(rules_fp)):,} rules")

if len(rules_ap) == len(rules_fp):
    print("   ‚úÖ HO√ÄN H·∫¢O! Gi·ªëng h·ªát nhau!")
elif abs(len(rules_ap) - len(rules_fp)) <= 5:
    print("   ‚úÖ G·∫¶N ƒê√öNG! Ch√™nh l·ªách < 5 rules (ch·∫•p nh·∫≠n ƒë∆∞·ª£c)")
else:
    print("   ‚ö†Ô∏è V·∫™N KH√ÅC NHAU - C·∫ßn ki·ªÉm tra th√™m")

# 2. So s√°nh support
print("\nüìä SO S√ÅNH MIN SUPPORT:")
print(f"Apriori min:   {rules_ap['support'].min():.6f}")
print(f"FP-Growth min: {rules_fp['support'].min():.6f}")

if abs(rules_ap['support'].min() - rules_fp['support'].min()) < 0.0001:
    print("   ‚úÖ MIN_SUPPORT gi·ªëng nhau!")
else:
    print("   ‚ö†Ô∏è V·∫´n kh√°c support threshold")

# 3. Th·ªëng k√™ chi ti·∫øt
print("\n" + "="*70)
print("üìà TH·ªêNG K√ä CHI TI·∫æT")
print("="*70)

print("\nAPRIORI:")
print(rules_ap[['support', 'confidence', 'lift']].describe().to_string())

print("\nFP-GROWTH:")
print(rules_fp[['support', 'confidence', 'lift']].describe().to_string())

# 4. Top 3 rules
print("\n" + "="*70)
print("üèÜ TOP 3 RULES THEO LIFT")
print("="*70)

print("\nApriori Top 3:")
if 'antecedents_str' in rules_ap.columns:
    cols = ['antecedents_str', 'consequents_str', 'lift']
else:
    cols = ['antecedents', 'consequents', 'lift']

print(rules_ap.nlargest(3, 'lift')[cols].to_string(index=False))

print("\nFP-Growth Top 3:")
print(rules_fp.nlargest(3, 'lift')[cols].to_string(index=False))

print("\n" + "="*70)
print("‚úÖ KI·ªÇM TRA HO√ÄN T·∫§T!")
print("="*70)
```
